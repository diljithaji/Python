Q1) Generate all prime numbers less than 100.Print them in a spiral format (like a clockwise
square spiral) of minimal size that fits all primes.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

def generate_primes(limit):
    primes = []
    for num in range(2, limit):
        if is_prime(num):
            primes.append(num)
    return primes

def create_spiral(primes):
    n = len(primes)
    

    size = 1
    while size * size < n:
        size += 1
    
    matrix = [[0 for _ in range(size)] for _ in range(size)]
    
    top, bottom, left, right = 0, size - 1, 0, size - 1
    index = 0
    
    while top <= bottom and left <= right and index < n:
        for col in range(left, right + 1):
            if index < n:
                matrix[top][col] = primes[index]
                index += 1
        top += 1
        
        for row in range(top, bottom + 1):
            if index < n:
                matrix[row][right] = primes[index]
                index += 1
        right -= 1
        
        if top <= bottom:
            for col in range(right, left - 1, -1):
                if index < n:
                    matrix[bottom][col] = primes[index]
                    index += 1
            bottom -= 1
        
        if left <= right:
            for row in range(bottom, top - 1, -1):
                if index < n:
                    matrix[row][left] = primes[index]
                    index += 1
            left += 1
    
    return matrix

def print_spiral(matrix):
    for row in matrix:
        for num in row:
            if num == 0:
                print("   ", end=" ")
            else:
                print(f"{num:3d}", end=" ")
        print()

primes = generate_primes(100)
print(f"Primes less than 100: {len(primes)} primes found")
print(f"\nPrimes: {primes}\n")
print("Spiral Format:")
spiral_matrix = create_spiral(primes)
print_spiral(spiral_matrix)
```

Q2) You are a data analyst at a streaming platform, managing user interactions, analytics,
and daily operations. Use a for loop to create a text-based graph of hourly views. For
example, use * to represent every 5 views.

```python

hourly_views = [45, 32, 28, 15, 12, 8, 10, 25, 50, 65, 78, 85,
                90, 88, 82, 95, 105, 110, 98, 87, 75, 60, 52, 40]

print("\nHourly Views (each * represents 5 views):\n")

for hour in range(len(hourly_views)):
    views = hourly_views[hour]
    stars = views // 5  
    
    hour_label = f"Hour {hour:2d}"
    
    bar = "*" * stars
    
    print(f"{hour_label} ({views:3d} views): {bar}")

print(f"\nTotal daily views: {sum(hourly_views)}")
```

Q3) Each video watched generates revenue based on ad impressions:
- First 10 views: $0.50 per view
- Next 20 views: $0.30 per view
- Remaining views: $0.10 per view
You are given hourly views for the day. Calculate total daily revenue.

```python
def calculate_revenue(views):
    revenue = 0.0
    
    if views <= 10:
        revenue = views * 0.50
    elif views <= 30:
        revenue = (10 * 0.50) + ((views - 10) * 0.30)
    else:
        revenue = (10 * 0.50) + (20 * 0.30) + ((views - 30) * 0.10)
    
    return revenue

hourly_views = [45, 32, 28, 15, 12, 8, 10, 25, 50, 65, 78, 85,90, 88, 82, 95, 105, 110, 98, 87, 75, 60, 52, 40]

total_revenue = 0.0

print("\nHourly Revenue Breakdown:")
print("-" * 60)

for hour in range(len(hourly_views)):
    views = hourly_views[hour]
    revenue = calculate_revenue(views)
    total_revenue += revenue
    print(f"Hour {hour:2d}: {views:3d} views → ${revenue:6.2f}")

print("-" * 60)
print(f"\nTotal Daily Views: {sum(hourly_views)}")
print(f"Total Daily Revenue: ${total_revenue:.2f}")

print("\nRevenue Tier Breakdown:")
total_views = sum(hourly_views)
tier1_views = min(total_views, 10)
tier2_views = min(max(total_views - 10, 0), 20)
tier3_views = max(total_views - 30, 0)

tier1_revenue = tier1_views * 0.50
tier2_revenue = tier2_views * 0.30
tier3_revenue = tier3_views * 0.10

print(f"  Tier 1 (first 10 views @ $0.50): {tier1_views:4d} views = ${tier1_revenue:7.2f}")
print(f"  Tier 2 (next 20 views @ $0.30): {tier2_views:4d} views = ${tier2_revenue:7.2f}")
print(f"  Tier 3 (remaining @ $0.10):     {tier3_views:4d} views = ${tier3_revenue:7.2f}")
print(f"  {'Total:':30s} {total_views:4d} views = ${tier1_revenue + tier2_revenue + tier3_revenue:7.2f}")
```
