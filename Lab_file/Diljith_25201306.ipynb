### A) Write a program to calculate area of a triangle using Heron’s Formula.

Date: 04/09/25

```python
import math

# Define the side lengths (input from user or hardcoded)
side_a = float(input("Enter side a: "))
side_b = float(input("Enter side b: "))
side_c = float(input("Enter side c: "))

print(f"Calculating area for sides: a={side_a}, b={side_b}, c={side_c}")

# Calculate the semi-perimeter
s = (side_a + side_b + side_c) / 2

# Calculate the area using Heron's formula
# Area = sqrt(s * (s-a) * (s-b) * (s-c))
area = math.sqrt(s * (s - side_a) * (s - side_b) * (s - side_c))

print(f"The area of the triangle is: {area:.2f}")
```

Explanation:This Python code calculates the area of a triangle using Heron's Formula. Here's a breakdown of what it does9:Input Side Lengths: It starts by asking the user to input three variables: side_a, side_b, and side_c.Calculating the Semi-perimeter (s): Heron's formula requires the semi-perimeter of the triangle, which is half of its perimeter.Applying Heron's Formula: The core calculation uses the math.sqrt function to evaluate $\sqrt{s(s-a)(s-b)(s-c)}$.

 Result: Finally, the calculated area is printed to the console, formatted to two decimal places using an f-string (:.2f)11.

### A) Write a program to solve the expression: x = 10 + 2 * 3 ** 2 - 8 / 4 and show each expression calculation and output.

Date: 18/09/25

```python
# Original expression
print("Original expression: x = 10 + 2 * 3 ** 2 - 8 / 4")

# Direct calculation
x = 10 + 2 * 3 ** 2 - 8 / 4
print(f"Final result: x = {x}")

print("\n--- Step-by-step Calculation ---")

# Step 1: Exponentiation (**) - Highest precedence
step1_val = 3 ** 2
print(f"1. Exponentiation (3**2): {step1_val}")
print(f"   Expression now: 10 + 2 * {step1_val} - 8 / 4")

# Step 2: Multiplication (*)
step2_val = 2 * step1_val
print(f"2. Multiplication (2 * {step1_val}): {step2_val}")
print(f"   Expression now: 10 + {step2_val} - 8 / 4")

# Step 3: Division (/)
step3_val = 8 / 4
print(f"3. Division (8 / 4): {step3_val}")
print(f"   Expression now: 10 + {step2_val} - {step3_val}")

# Step 4: Addition (+)
step4_val = 10 + step2_val
print(f"4. Addition (10 + {step2_val}): {step4_val}")
print(f"   Expression now: {step4_val} - {step3_val}")

# Step 5: Subtraction (-)
step5_val = step4_val - step3_val
print(f"5. Subtraction ({step4_val} - {step3_val}): {step5_val}")

# Verification
print(f"\nVerification with Python's direct calculation: {x}")
```

Explanation: This code demonstrates how Python evaluates the expression x = 10 + 2 * 3 ** 2 - 8 / 4.

Original Expression: It calculates the full expression directly to get the final result 26.0.

Step-by-step Breakdown: The program breaks down the calculation adhering to Python's operator precedence (PEMDAS/BODMAS):


Exponentiation: 3 ** 2 is evaluated first (9). 


Multiplication/Division: 2 * 9 (18) and 8 / 4 (2.0) are evaluated next. 


Addition/Subtraction: Finally, 10 + 18 (28) and 28 - 2.0 (26.0) are calculated.

### C) Write a program to print the reverse of a number.

Date: 18/09/25

```python
num = int(input("Enter an integer: "))
original_num = num
reversed_num = 0

while num > 0:
    digit = num % 10
    reversed_num = reversed_num * 10 + digit
    num = num // 10

print(f"Original: {original_num}")
print(f"Reversed: {reversed_num}")
```

Explanation: This program reverses an integer mathematically using a while loop. 


Modulo: num % 10 extracts the last digit. 


Build Reverse: reversed_num * 10 + digit shifts the current reverse value to the left and adds the new digit. 

Floor Division: num // 10 removes the last digit from the original number. This repeats until num becomes 0.

### B) Write a recursive function count_digits(n) that returns the number of digits.

Date: 26/09/25

```python
def count_digits(n):
    if n < 10:
        return 1
    return 1 + count_digits(n // 10)

n = int(input("Enter number: "))
print(f"Digits in {n}: {count_digits(n)}")
```

Explanation: This uses recursion to count digits. 


Base Case: If n < 10, it returns 1 (single digit). 


Recursive Step: Otherwise, it adds 1 to the result of count_digits(n // 10), effectively stripping one digit and counting it.

### B) Check for palindrome ignoring case and symbols.
Date: 16/10/25

```python
s = input("Enter string: ")
# Keep only alphanumeric and make lowercase
clean_s = "".join(ch.lower() for ch in s if ch.isalnum())

if clean_s == clean_s[::-1]:
    print("It is a palindrome.")
else:
    print("Not a palindrome.")
```

Explanation:

Comparison: clean_s[::-1] reverses the string. If the cleaned string matches its reverse, it is a palindrome.

### A) Overload + operator for Student class.

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __add__(self, other):
        return self.marks + other.marks

s1 = Student("A", 80)
s2 = Student("B", 90)
print(f"Total Marks: {s1 + s2}")
```
